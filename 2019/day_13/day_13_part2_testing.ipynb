#### assumptions: 

- we need inputs now:
    - if the ball is to the left, we go left. 
    - if the ball is to the right, we go right....


```python
# initial intermed_dict vals
intermed_dict = {}
intermed_dict['list'] = lst
intermed_dict['init_val'] = 0
intermed_dict['rel_base'] = 0
intermed_dict['robo_input'] = 0
input_val = 0 # start neutral? 

# assume ball & paddle start at (0,0)?
ball_loc = 0
pad_loc = 0
```

```python
import sys
game_dict = {}
j = 0
while True:
    
    # empty list
    empty_list = []
    
    # run 3 times to store each output
    for i in range(3):
        # run process, updating
        intermed_dict = intcode_pc(list_vals = intermed_dict['list'], init_val = intermed_dict['init_val'], rel_base=intermed_dict['rel_base'] , input_val = input_val)
        empty_list.append(intermed_dict['robo_output'])
        
        # Find our stopping point
        if intermed_dict['robo_output'] == 'FINISHED':
            sys.exit() # trying to kill the process...
            
    # break this out into proper coords & val
    pos_x = empty_list[0]
    pos_y = empty_list[1]
    tile_id = empty_list[2]
    
    # print score
    if pos_x == -1 and pos_y == 0:
        print(f"Your score is currently: {tile_id}")
              
    # find paddle x-loc
    if tile_id == 3:
        pad_loc = pos_x
              
    # find ball x-loc
    if tile_id == 4:
        ball_loc = pos_x
    
    # can't do anything if we have Nones...
    if ball_loc is None or pad_loc is None:
        input_val = 0
        
    # determine input:if a ball we can move
    elif tile_id == 4:
        # if ball x-pos less than paddle y-pos, move left
        if ball_loc < pad_loc:
            input_val = -1

        # elif ball x-pos > than paddle y-pos, move right
        elif ball_loc > pad_loc:
            input_val = 1
        
        # if equal
        else:
            input_val = 0
    else:
        input_val = 0
    
    # add to dict
    game_dict[(pos_x, pos_y)] = tile_id

    
```
