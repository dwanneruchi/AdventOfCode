```python
# empty list
data_list = []

# need to ensure a clean run each time
with open('day12_data.txt') as f:
    for line in f:
        data_list.append(line.strip('\n').split(' '))

# planet data
planet_data = {}

# intermediat step info
planet_int_data = {}

for i in range(len(data_list)):
   
   
    pos_list = [int(x.split('=')[1].strip(',').strip('>')) for x in data_list[i]]
    vel_list = [0,0,0]
   
    planet_int_data[i] = [pos_list, vel_list]
# add to step 0

planet_data[0] = planet_int_data

### Now we can build a functiojn to output velocity

def grav_vel(val1, val2):
   
    if val1 == val2:
        vel = 0
    elif val1 > val2:
        vel = -1
    else:
        vel = 1
   
    return vel

# Energy funcs

def p_energy(x,y,z):
   
    return abs(x) + abs(y) + abs(z)

def k_energy(x,y,z):
   
    return abs(x) + abs(y) + abs(z)

def tot_energy(p_energy, k_energy):
   
    return p_energy * k_energy


# while loop
step = 1

while step < 1001:
    step_prev = step - 1
    planet_int_data = {}

    for i in range(len(data_list)):
       
        # all planets except i:
        other_planets = [x for x in range(len(data_list)) if x != i]
       
        # find previous vals of this planet
        x,y,z = tuple(planet_data[step_prev][i][0])
        x_v, y_v, z_v = tuple(planet_data[step_prev][i][1])
       
        # build new velocity
        xv_new, yv_new, zv_new = 0, 0, 0
       
        for j in other_planets:

            # capture other planets x,y,z
            x_o, y_o, z_o = tuple(planet_data[step_prev][j][0])
           
            # calculate velocities
            xv_new += grav_vel(x,x_o)
            yv_new += grav_vel(y,y_o)
            zv_new += grav_vel(z,z_o)
           
        # add new velocities to list:
        vel_list = [x_v + xv_new, y_v + yv_new, z_v + zv_new]
       
        # use velocities to calculate change in position
        pos_list = [x + vel_list[0], y + vel_list[1], z + vel_list[2]]
   
        planet_int_data[i] = [pos_list, vel_list]
   
    # add to dict under "step"
    planet_data[step] = planet_int_data
   
    # moveforward
    step += 1


# steps
steps = 1000

# all energy:
all_energy = 0

for planet in planet_data[steps].keys():
   
    # find positions
    x,y,z = tuple(planet_data[steps][planet][0])
   
    # find velocities
    xv, yv, zv = tuple(planet_data[steps][planet][1])
   
    # calculate energy
    k_e = k_energy(x,y,z)
    p_e = p_energy(xv, yv, zv)
    tot_e = tot_energy(k_e, p_e)
   
    # add to all energy:
    all_energy += tot_e
```
