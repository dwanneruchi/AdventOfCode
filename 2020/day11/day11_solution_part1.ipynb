```python
from copy import deepcopy

filepath = "day11_test_data.txt"
with open(filepath) as fh:
    lines = [line.strip() for line in fh.readlines()]

lines = [x for x in lines]
test_list = []
for row in lines:
    test_list.append([x for x in row])

lines = test_list
    
```

```python


row_size = len(lines)
col_size = len(lines[0])


for _ in range(6):
    state_change = 0
    output_list = deepcopy(lines)

    for row in range(len(lines)):
        for col in range(len(lines[0])):
            #print(f"Checking out {row, col}")
            starting_state = lines[row][col]

            # get adjacencies 
            check_space = [(row, col - 1), 
                   (row - 1, col - 1), 
                   (row - 1, col), 
                   (row - 1, col + 1), 
                   (row, col + 1),
                   (row + 1, col + 1), 
                   (row + 1, col), 
                   (row + 1, col - 1),              
                  ]

            # remove any rows or cols that are outside the bounds allowed
            final_space = [(s[0], s[1]) for s in check_space if 0 <= s[0] < row_size and  0 <= s[1] < col_size]

            if lines[row][col] == 'L':
                open_seat = True
                for r, c in final_space:
                    if lines[r][c] == '#':
                        output_list[row][col] = 'L'
                        open_seat = False
                        
                if open_seat:
                    output_list[row][col] = '#'



            elif lines[row][col] == '#':

                #print(final_space)

                # determine how many other seats are occupied:
                count = 0
                for r, c in final_space:
                    #print(lines[r][c])
                    if lines[r][c] == '#':
                        count += 1
                if count >= 4:
                    output_list[row][col] = 'L'
                else:
                    output_list[row][col] = '#'
            else:
                output_list[row][col] = '.'
                continue

            #print("Tested")
            if output_list[row][col] != starting_state:
                state_change += 1
    # this is so ugly
    lines = deepcopy(output_list)
    
    if state_change < 1:
        break
    
    print(f"Iteration: {_}")
    print(f"Total State Changes: {state_change}")
    print(lines)

# Total occupied seats:
occ_seats = 0
for row in range(len(lines)):
    for col in range(len(lines[0])):
        if lines[row][col] == "#":
            occ_seats += 1
print(f"Total Occupied: {occ_seats}")   
        
```
